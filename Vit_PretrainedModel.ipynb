```python
from transformers import ViTForImageClassification, ViTFeatureExtractor

# Load pre-trained Vision Transformer model
model_name = "google/vit-base-patch16-224"
model = ViTForImageClassification.from_pretrained(model_name)
feature_extractor = ViTFeatureExtractor.from_pretrained(model_name)

print("Model and Feature Extractor Loaded!")
```

```python
# Save the model locally
model.save_pretrained("./vit_model")
feature_extractor.save_pretrained("./vit_model")

# Load the model from local directory
model = ViTForImageClassification.from_pretrained("./vit_model")
feature_extractor = ViTFeatureExtractor.from_pretrained("./vit_model")
```
